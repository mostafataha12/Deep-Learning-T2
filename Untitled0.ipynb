```python
import math
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
X = [1, 2, 3]
print("Input sequence:", X)
```

```python
Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0

Wy, by = 4, 0
```

```python
h = 0
C = 0

print("Initial hidden state h0 =", h)
print("Initial cell state C0 =", C)
```

```python
for t, x in enumerate(X, start=1):
    f = sigmoid(Wf * x + Whf * h + bf)
    i = sigmoid(Wi * x + Whi * h + bi)
    C_tilde = math.tanh(Wc * x + Whc * h + bc)
    C = f * C + i * C_tilde
    o = sigmoid(Wo * x + Who * h + bo)
    h = o * math.tanh(C)

    print(f"Time Step t = {t}, Input x{t} = {x}")
    print(f"Forget gate f{t}      = {f:.3f}")
    print(f"Input gate i{t}       = {i:.3f}")
    print(f"Candidate state C~{t} = {C_tilde:.3f}")
    print(f"Cell state C{t}       = {C:.3f}")
    print(f"Output gate o{t}      = {o:.3f}")
    print(f"Hidden state h{t}     = {h:.3f}")
    print("-" * 40)
```

```python
x4 = 4

f4 = sigmoid(Wf * x4 + Whf * h + bf)
i4 = sigmoid(Wi * x4 + Whi * h + bi)
C_tilde4 = math.tanh(Wc * x4 + Whc * h + bc)
C4 = f4 * C + i4 * C_tilde4
o4 = sigmoid(Wo * x4 + Who * h + bo)
h4 = o4 * math.tanh(C4)

print("Prediction step for x4")
print(f"Forget gate f4      = {f4:.3f}")
print(f"Input gate i4       = {i4:.3f}")
print(f"Candidate state C~4 = {C_tilde4:.3f}")
print(f"Cell state C4       = {C4:.3f}")
print(f"Output gate o4      = {o4:.3f}")
print(f"Hidden state h4     = {h4:.3f}")
```

```python
y_hat = Wy * h4 + by

print("Final Output Layer")
print(f"y_hat = {Wy} * {h4:.3f} + {by}")
print(f"Predicted next value = {y_hat:.3f}")
```
